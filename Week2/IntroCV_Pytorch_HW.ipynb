````python
import torch
from torch import nn


# Function
class Function_A(nn.Module):
    def __init__(self):
        super().__init__()
        print(f"        Function A Initialized")

    def forward(self, x):
        print(f"        Function A started")
        print(f"        Function A done")

class Function_B(nn.Module):
    def __init__(self):
        super().__init__()
        print(f"        Function B Initialized")

    def forward(self, x):
        print(f"        Function B started")
        print(f"        Function B done")

class Function_C(nn.Module):
    def __init__(self):
        super().__init__()
        print(f"        Function C Initialized")

    def forward(self, x):
        print(f"        Function C started")
        print(f"        Function C done")

class Function_D(nn.Module):
    def __init__(self):
        super().__init__()
        print(f"        Function D Initialized")

    def forward(self, x):
        print(f"        Function D started")
        print(f"        Function D done")


# Layer
class Layer_AB(nn.Module):
    def __init__(self):
        super().__init__()

        self.a = Function_A()
        self.b = Function_B()

        print(f"    Layer AB Initialized")

    def forward(self, x):
        print(f"    Layer AB started")
        self.a(x)
        self.b(x)
        print(f"    Layer AB done")

class Layer_CD(nn.Module):
    def __init__(self):
        super().__init__()

        self.c = Function_C()
        self.d = Function_D()

        print(f"    Layer CD Initialized")

    def forward(self, x):
        print(f"    Layer CD started")
        self.c(x)
        self.d(x)
        print(f"    Layer CD done")


# Model
class Model(nn.Module):
    def __init__(self):
        super().__init__()

        self.ab = Layer_AB()
        self.cd = Layer_CD()

        print(f"Model ABCD Initialized\n")

    def forward(self, x):
        print(f"Model ABCD started")
        self.ab(x)
        self.cd(x)
        print(f"Model ABCD done\n")


x = torch.tensor([7])

model = Model()
model(x)

print("ğŸ‰ğŸ‰ğŸ‰ ëª¨ë“  ë”¥ëŸ¬ë‹ ëª¨ë¸ì€ ì´ì²˜ëŸ¼ Moduleë“¤ì´ ìŒ“ì´ê³  ìŒ“ì—¬ì„œ ë§Œë“¤ì–´ì§‘ë‹ˆë‹¤! ğŸ‰ğŸ‰ğŸ‰")
print("ğŸ‰ğŸ‰ğŸ‰ íë¦„ì„ ëŠê»´ë³´ì‹œê³  ì´ íë¦„ì´ ì´í•´ê°€ ë˜ì‹  ë¶„ì€ ë‹¤ìŒìœ¼ë¡œ ê°€ì‹œë©´ ë©ë‹ˆë‹¤! ğŸ‰ğŸ‰")
````

````python
import torch
from sklearn.datasets import load_iris
from torch.utils.data import Dataset

class IrisDataset(Dataset):
    def __init__(self):
        iris = load_iris()

        self.X = iris['data']
        self.y = iris['target']
        self.feature_names = iris['feature_names']
        self.target_names = iris['target_names']

    def __len__(self):
        len_dataset = len(self.X)
        return len_dataset

    def __getitem__(self, idx):
        X = torch.tensor(self.X[idx], dtype=torch.float)
        y = torch.tensor(self.y[idx], dtype=torch.long)
        return X, y
````

```python
(3) ì•ì„œ ë³€í™˜ëœ ì´ë¯¸ì§€ì—ì„œ ì¤‘ê°„ ë¶€ë¶„ì„ ê°€ë¡œ, ì„¸ë¡œë¥¼ 150ë§Œí¼ ì˜ë¼ë‚´ì–´ ì£¼ì„¸ìš”.(CenterCrop)
```

````python
def get_transforms_img(im):
    # transform = transforms.Resize((224,224))(im)
    resize_transform = transforms.Compose([
    transforms.Resize((224, 224))
    ])

    resized_image = resize_transform(im)


    flip_transform = transforms.Compose([
    transforms.RandomVerticalFlip(p=1)
    ])

    flipped_image = flip_transform(resized_image)

    crop_transform = transforms.Compose([
    transforms.CenterCrop(150)
    ])

    cropped_image = crop_transform(flipped_image)


    return cropped_image
````

````python
get_transforms_img(im)
````
